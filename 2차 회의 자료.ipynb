- aliexpress, amazon 모두 같은 사람이 여러번 리뷰를 남긴 경우는 없는 것으로 보인다!

```python
print(alibaba['thumbsUpCount'].describe())
print(aliexpress['thumbsUpCount'].describe())
print(amazon['thumbsUpCount'].describe())
```

```python
plt.hist(alibaba['thumbsUpCount'], alpha=0.5, color='red',bins=10000, label='alibaba')
plt.xlim(0, 100)
plt.xlabel('Thumbs Up Count')
plt.title('Histogram of Alibaba Thumbs Up Count')
plt.legend()
plt.show()
```

```python
print('10 이하 갯수 : ', (alibaba['thumbsUpCount'] <= 10).sum())
print('10 초과 50 이하 갯수 : ', ((alibaba['thumbsUpCount'] > 10)&(alibaba['thumbsUpCount']<=50)).sum())
print('50 초과 100 이하 갯수 : ', ((alibaba['thumbsUpCount'] > 20)&(alibaba['thumbsUpCount']<=100)).sum())
print('100 초과 갯수 : ', ((alibaba['thumbsUpCount'] > 100)).sum())
```

```python
plt.hist(aliexpress['thumbsUpCount'], alpha=0.5, color='green', bins=1000, label='aliexpress', density=True)
plt.xlim(0, 1000)
plt.xlabel('Thumbs Up Count')
plt.title('Histogram of Aliexpress Thumbs Up Count')
plt.legend()
plt.show()
```

```python
print('10 이하 갯수 : ', (aliexpress['thumbsUpCount'] <= 10).sum())
print('10 초과 20 이하 갯수 : ', ((aliexpress['thumbsUpCount'] > 10)&(aliexpress['thumbsUpCount']<=20)).sum())
print('20 초과 50 이하 갯수 : ', ((aliexpress['thumbsUpCount'] > 20)&(aliexpress['thumbsUpCount']<=50)).sum())
print('50 초과 갯수 : ', ((aliexpress['thumbsUpCount'] > 50)).sum())
```

```python
plt.hist(amazon['thumbsUpCount'], alpha=0.5, color='blue', bins=10000, label='aliexpress', density=True)
plt.xlim(0, 100)
plt.xlabel('Thumbs Up Count')
plt.title('Histogram of Aliexpress Thumbs Up Count')
plt.legend()
plt.show()
```

```python
print('10 이하 갯수 : ', (amazon['thumbsUpCount'] <= 10).sum())
print('10 초과 50 이하 갯수 : ', ((amazon['thumbsUpCount'] > 10)&(amazon['thumbsUpCount']<=50)).sum())
print('50 초과 100 이하 갯수 : ', ((amazon['thumbsUpCount'] > 50)&(amazon['thumbsUpCount']<=100)).sum())
print('100 초과 갯수 : ', ((amazon['thumbsUpCount'] > 100)).sum())
```
